```python
import requests
import pandas as pd
import datetime
import time
from concurrent import futures
```

```python
time_range = pd.date_range(start='2019-01-01 00:00:00', end='2019-01-02 00:00:00', freq='min')
time_range
```

```python
time_range[0].strftime('%Y-%m-%dT%H%%3A%M%%3A%S')
```

```python
json_list = []

base_url = 'https://api.data.gov.sg/v1/transport/traffic-images?date_time='

for t in time_range:
  temp_url = base_url + t.strftime('%Y-%m-%dT%H%%3A%M%%3A%S')
  r = requests.get(temp_url, auth=('user', 'pass'))
  r_json = r.json()
  json_list.append(r_json)
```

```python
df_list = []
for json in json_list:
  df_list.append(pd.json_normalize(json['items'][0]['cameras']))
df_cameras = pd.concat(df_list)
df_cameras
```

```python
df_cameras.reset_index(drop=True, inplace=True)
df_cameras
```

```python
df_cameras[df_cameras['camera_id'] == '1709']
```

```python
df_cameras.drop_duplicates(subset=['image_metadata.md5'], inplace=True)
df_cameras
```

```python
time_range_2019 = pd.date_range(start='2019-01-01 00:00:00', end='2019-02-01 00:00:00', freq='min')
time_range_2020 = pd.date_range(start='2020-01-01 00:00:00', end='2020-02-01 00:00:00', freq='min')
time_range_2019
```

```python
def retrive_camera_json(url):
  base_url = 'https://api.data.gov.sg/v1/transport/traffic-images?date_time='
  temp_url = base_url + url.strftime('%Y-%m-%dT%H%%3A%M%%3A%S')
  
  r = requests.get(temp_url, auth=('user', 'pass'))
  r_json = r.json()
  
  return r_json
```

```python
future_results = []

with futures.ThreadPoolExecutor(max_workers=150) as executor:
  futures_json = [executor.submit(retrive_camera_json, url) for url in time_range_2019]
  for future in futures.as_completed(futures_json):
    future_results.append(future.result()) 
```

```python
df_list = []
for json in future_results:
  df_list.append(pd.json_normalize(json['items'][0]['cameras']))
df_cameras_2019 = pd.concat(df_list, ignore_index=True)
df_cameras_2019
```

```python
df_cameras_2019.drop_duplicates(subset=['image_metadata.md5'], inplace=True)
df_cameras_2019
```

```python
df_cameras_2019.sort_values(by=['timestamp'], ignore_index=True, inplace=True)
df_cameras_2019
```

```python
df_cameras_2019.to_csv('/content/drive/MyDrive/Colab Notebooks/Project Traffic/traffic_2019.csv')
```
