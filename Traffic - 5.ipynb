```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_2019 = pd.read_csv('traffic_2019_with_car_nums.csv', parse_dates=['timestamp'], index_col = 1)
df_2019
```

```python
df_2019.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_2019.info()
```

```python
df_2020 = pd.read_csv('traffic_2020_with_car_nums.csv', parse_dates=['timestamp'], index_col = 1)
df_2020
```

```python
df_2020.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_2020.info()
```

```python
sns.histplot(data=df_2019, x='num_cars')
plt.title('Number of cars 2019')
plt.show()
```

```python
sns.histplot(data=df_2020, x='num_cars')
plt.title('Number of cars 2020')
plt.show()
```

```python
sns.histplot(data=df_2019, x='num_cars', color = 'blue', alpha=0.5, kde=True)
sns.histplot(data=df_2020, x='num_cars', color = 'green', alpha=0.5, kde=True)
plt.title('Number of cars for both years')
plt.show()
```

```python
plt.hist([df_2019['num_cars'], df_2020['num_cars']], color=['r', 'b'], alpha=0.5, log=True)
plt.title('Log-scaled number of cars for both years')
plt.show()
```

```python
resampled_by_day_2019 = df_2019.resample('D')['num_cars'].sum()
resampled_by_day_2019
```

```python
resampled_by_day_2020 = df_2020.resample('D')['num_cars'].sum()
resampled_by_day_2020
```

```python
plt.plot(resampled_by_day_2020.values[1:], label='2020')
plt.plot(resampled_by_day_2019.values[1:], label='2019')
plt.title('Numbers of cars per day for both years')
plt.legend()
plt.show()
```

```python
df_2019['day_of_week'] = df_2019.index.dayofweek
df_2019
```

```python
df_2020['day_of_week'] = df_2020.index.dayofweek
df_2020
```

```python
df_2019.drop(index=df_2019.index[0],
            axis=0,
            inplace=True)
df_2019
```

```python
df_2020.drop(index=df_2020.index[0],
            axis=0,
            inplace=True)
df_2020
```

```python
groupby_2019 = df_2019.groupby(by=['day_of_week'])['num_cars'].mean()
groupby_2019
```

```python
groupby_2020 = df_2020.groupby(by=['day_of_week'])['num_cars'].mean()
groupby_2020
```

```python
plt.plot(groupby_2019, label='2019')
plt.plot(groupby_2020, label='2020')
plt.legend()
plt.title('Average traffic per day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Avg cars')
plt.show()
```

```python
groupby_2019 = df_2019[df_2019['num_cars'] > 0].groupby(by=['day_of_week'])['num_cars'].mean()
groupby_2019
```

```python
groupby_2020 = df_2020[df_2020['num_cars'] > 0].groupby(by=['day_of_week'])['num_cars'].mean()
groupby_2020
```

```python
plt.plot(groupby_2019, label='2019')
plt.plot(groupby_2020, label='2020')
plt.legend()
plt.title('Average traffic per day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Avg cars')
plt.show()
```
